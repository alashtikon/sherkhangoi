лњлЙ-л▓ЛѓлЙЛђЛІЛЁ, л╝лЙлХлхл╝ лИЛЂл┐лЙл╗ЛїлилЙл▓л░ЛѓЛї л┤лЙл╗л│лЙЛѓЛЃ лИ ЛѕлИЛђлЙЛѓЛЃ ЛѓлЙЛЄлхл║ ЛЂЛѓл░ЛђЛѓл░/лил░л▓лхЛђЛѕлхлйлИЛЈ л┐лЙлхлил┤л║лИ, ЛЄЛѓлЙл▒ЛІ л┐ЛђлИл╝лхЛђлйлЙ лЙЛєлхлйлИЛѓЛї Лђл░ЛЂЛЂЛѓлЙЛЈлйлИлх л╝лхлХл┤ЛЃ 2 ЛѓлЙЛЄл║л░л╝лИ.

лАл░л╝лИ л┐лЙ ЛЂлхл▒лх лЙлйлИ, л║л░л║ ЛЂл░л╝лЙЛЂЛѓлЙЛЈЛѓлхл╗ЛїлйЛІлх л▓лхЛЅлхЛЂЛѓл▓лхлйлйЛІлх л┐ЛђлИлилйл░л║лИ, л▓ЛђЛЈл┤ л╗лИ ЛЂл┐лЙЛЂлЙл▒лйЛІ ЛЁлЙЛђлЙЛѕлЙ лЙл▒ЛіЛЈЛЂлйЛЈЛѓЛї л┤л╗лИЛѓлхл╗ЛїлйлЙЛЂЛѓЛї л┐лЙлхлил┤л║лИ.

лЉл░лилЙл▓л░ЛЈ лИл┤лхЛЈ ЛЂлЙЛЂЛѓлЙлИЛѓ л▓ ЛѓлЙл╝, ЛЄЛѓлЙл▒ЛІ л┐лЙЛЂЛЄлИЛѓл░ЛѓЛї Лђл░лилйлЙЛЂЛѓЛї л┤лЙл╗л│лЙЛѓ лИ ЛѕлИЛђлЙЛѓ ЛЂлЙлЙЛѓл▓лхЛѓЛЂЛѓл▓лхлйлйлЙ, ЛѓлЙ лхЛЂЛѓЛї:

$$
\delta_{long} = \text{dropoff_longitude} - \text{pickup_longitude}
$$

$$
\delta_{lat} =  \text{dropoff_latitude} - \text{pickup_latitude}
$$

лљ л┐лЙЛѓлЙл╝ л▓ЛІЛЄлИЛЂл╗лИЛѓЛї л│лхлЙл│Лђл░ЛёлИЛЄлхЛЂл║лЙлх Лђл░ЛЂЛЂЛѓлЙЛЈлйлИлх л╝лхлХл┤ЛЃ 2 ЛѓлЙЛЄл║л░л╝лИ л┐лЙ ЛѓлхлЙЛђлхл╝лх лЪлИЛёл░л│лЙЛђл░:

$$
R = \sqrt{\delta^2_{long} + \delta^2_{lat}}
$$

люЛІ Лђлхл░л╗лИлиЛЃлхл╝ л┤л░лйлйЛЃЛј лил░л┤ЛЃл╝л║ЛЃ лИ л▓ЛІЛЄлИЛЂл╗лИл╝ Лѓл░л║ЛЃЛј л▓лхЛЅлхЛЂЛѓл▓лхлйлйЛЃЛј л║лЙл╗лЙлйл║ЛЃ **R**, ЛЄЛѓлЙ, л▓ Лєлхл╗лЙл╝, ЛЈл▓л╗ЛЈлхЛѓЛЂЛЈ ЛЁлЙЛђлЙЛѕлИл╝ ЛѓлЙлйлЙл╝ л┐ЛђлИ Лђл░л▒лЙЛѓлх ЛЂ л║лЙлЙЛђл┤лИлйл░Лѓл░л╝лИ ЛѓлЙЛЄлхл║.

лблЙл╗Лїл║лЙ л┤л╗ЛЈ лйл░ЛЄл░л╗л░ лйЛЃлХлйлЙ лйлхл║лЙЛѓлЙЛђЛІл╝ лЙл▒Лђл░лилЙл╝ л┐лхЛђлхл▓лхЛЂЛѓлИ л┤лЙл╗л│лЙЛѓЛЃ лИ ЛѕлИЛђлЙЛѓЛЃ л▓ л║лИл╗лЙл╝лхЛѓЛђЛІ, лЙл▒лхЛЂл┐лхЛЄлИл▓ Лђл░л▓лхлйЛЂЛѓл▓лЙ лИЛЁ л╝лхЛђ лИлил╝лхЛђлхлйлИЛЈ. лЪлЙЛѓлЙл╝ЛЃ ЛЄЛѓлЙ, л▓лЙлЙл▒ЛЅлх л│лЙл▓лЙЛђЛЈ, *л│Лђл░л┤ЛЃЛЂлйл░ЛЈ л╝лхЛђл░* ЛѕлИЛђлЙЛѓ лИ л┤лЙл╗л│лЙЛѓ лИл╝лхлхЛѓ лйлхлЙл┤лИлйл░л║лЙл▓ЛЃЛј Лѕл║л░л╗ЛЃ л┐лхЛђлхл▓лЙл┤л░ л▓ л║лИл╗лЙл╝лхЛѓЛђЛІ. лбл░л║, лхЛЂл╗лИ л┐ЛђлЙл┐ЛЃЛЂЛѓлИЛѓЛї л┤л░лйлйЛЃЛј л┤лхЛѓл░л╗Лї, Лђл░ЛЂЛЂЛѓлЙЛЈлйлИлх **R** л▒ЛЃл┤лхЛѓ л▓ЛІЛЄлИЛЂл╗лхлйлЙ лйлхл▓лхЛђлйлЙ, л▓лхл┤Лї л║л░ЛѓлхЛѓЛІ ЛѓлЙл│л┤л░ л▒ЛЃл┤ЛЃЛѓ лИл╝лхЛѓЛї Лђл░лилйЛЃЛј Лђл░лил╝лхЛђлйлЙЛЂЛѓЛї.

лњ Лєлхл╗лЙл╝, л┐лхЛђлхл▓лЙл┤ лИли л┤лЙл╗л│лЙЛѓ лИ ЛѕлИЛђлЙЛѓ л▓ Лђл░ЛЂЛЂЛѓлЙЛЈлйлИлх л┐лЙлхлил┤л║лИ л┐лЙлил▓лЙл╗ЛЈлхЛѓ лйл░л╝ л▓ л▒ЛЃл┤ЛЃЛЅлхл╝ л┐ЛђлЙл▓лхЛђлИЛѓЛї лил░л▓лИЛЂлИл╝лЙЛЂЛѓЛї **л┤л╗лИЛѓлхл╗ЛїлйлЙЛЂЛѓлИ л┐лЙлхлил┤л║лИ лЙЛѓ л║лИл╗лЙл╝лхЛѓЛђл░лХл░**, лИ лЙл▒ЛіЛЈЛЂлйлИЛѓЛї лхлх л▒ЛЃл┤лхЛѓ л║ЛЃл┤л░ л┐ЛђлЙЛЅлх, ЛЄлхл╝ л░лйл░л╗лЙл│лИЛЄлйЛЃЛј л╝лхлХл┤ЛЃ Лѓл░Лђл│лхЛѓлЙл╝ лИ лИлилйл░ЛЄл░л╗ЛїлйЛІл╝лИ л┐ЛђлИлилйл░л║л░л╝лИ

лџл░л║лЙл╣ л┐лЙл┤ЛЁлЙл┤ л▓ЛІл▒Лђл░ЛѓЛї л╗ЛЃЛЄЛѕлх лил░Лђл░лйлхлх лйл░л▓лхЛђлйЛЈл║л░ лйлх ЛЃлилйл░лхЛѕЛї. лЮЛЃлХлйЛІ ЛЇл║ЛЂл┐лхЛђлИл╝лхлйЛѓЛІ ЛЂ л┤л░лйлйЛІл╝лИ лИ л╝лЙл┤лхл╗ЛЈл╝лИ. лблхл╝ лйлх л╝лхлйлхлх, ЛЈ л┐Лђлхл┤л╗л░л│л░Лј лњл░л╝ л┐Лђлхл┤л┐лЙл╗лЙлХлИЛѓЛї, ЛЄЛѓлЙ л┤л░лйлйЛІл╣ л┐ЛђлИлилйл░л║ ЛЈл▓л╗ЛЈлхЛѓЛЂЛЈ л║л░Лѓлхл│лЙЛђлИл░л╗ЛїлйЛІл╝, лИ л┐лЙл┐ЛђлЙл▒лЙл▓л░ЛѓЛї лЙЛѓЛѓлЙЛЄлИЛѓЛї лйл░л▓ЛІл║лИ л║лЙл┤лИЛђлЙл▓л║лИ Лѓл░л║лИЛЁ ЛёлИЛЄлхл╣!

лЪЛђлхл┤л╗л░л│л░Лј лњл░л╝ Лђлхл░л╗лИлилЙл▓л░ЛѓЛї л┐ЛђлИлхл╝ ЛЂ **Mean-target encoding'лЙл╝**, л║л░л║ л▓ л┐Лђл░л║ЛѓлИЛЄлхЛЂл║лЙл╝ лил░лйЛЈЛѓлИлИ. лЌл░л╝лхлйлИЛѓлх л║лЙл╗лЙлйл║ЛЃ **passenger_count** л║лЙл╗лЙлйл║лЙл╣ **category_encoded**.
